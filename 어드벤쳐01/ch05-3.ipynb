```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap='gray')
```

```python
outimg = cv2.GaussianBlur(gimg, (3,3), 1)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (3,3), 3)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (7,7), 1)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (7,7), 3)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (15,15), 1)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg, (15,15), 3)
cv2_imshow(outimg)
```

```python
mean_img = cv2.blur(gimg, (5,5))
cv2_imshow(mean_img)
```

```python
edge_img = cv2.addWeighted(gimg, 1.0, mean_img, -1.0, 0)
cv2_imshow(edge_img)
```

```python
out_img = cv2.addWeighted(gimg, 1.0, edge_img, 1.0, 0)
cv2_imshow(out_img)
```

```python
out_img = cv2.addWeighted(gimg, 1.0, edge_img, 3.0, 0)
cv2_imshow(out_img)
```
