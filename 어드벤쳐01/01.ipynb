```python
import matplotlib.pyplot as plt
import networkx as nx

# 그래프 생성
G = nx.Graph()

# 중심 주제 노드 추가
G.add_node("주제")

# 서브 주제와 하위 주제들 추가
subtopics_1 = ["서브주제 1", "서브주제 2", "서브주제 3"]
subtopics_2 = ["서브주제 4", "서브주제 5", "서브주제 6"]
subtopics_3 = ["서브주제 7", "서브주제 8", "서브주제 9"]
subtopics_4 = ["서브주제 10", "서브주제 11"]

# 서브주제들 추가
G.add_nodes_from(subtopics_1)
G.add_nodes_from(subtopics_2)
G.add_nodes_from(subtopics_3)
G.add_nodes_from(subtopics_4)

# 부모-자식 노드 연결
for sub in subtopics_1:
    G.add_edge("주제", sub)
for sub in subtopics_2:
    G.add_edge("주제", sub)
for sub in subtopics_3:
    G.add_edge("주제", sub)
for sub in subtopics_4:
    G.add_edge("주제", sub)

# 각 서브 주제에서 추가 하위 주제들 연결
G.add_edge("서브주제 1", "하위주제 1.1")
G.add_edge("서브주제 1", "하위주제 1.2")
G.add_edge("서브주제 2", "하위주제 2.1")
G.add_edge("서브주제 2", "하위주제 2.2")
G.add_edge("서브주제 3", "하위주제 3.1")
G.add_edge("서브주제 3", "하위주제 3.2")
G.add_edge("서브주제 4", "하위주제 4.1")
G.add_edge("서브주제 5", "하위주제 5.1")
G.add_edge("서브주제 6", "하위주제 6.1")
G.add_edge("서브주제 6", "하위주제 6.2")
G.add_edge("서브주제 7", "하위주제 7.1")
G.add_edge("서브주제 8", "하위주제 8.1")
G.add_edge("서브주제 9", "하위주제 9.1")
G.add_edge("서브주제 10", "하위주제 10.1")
G.add_edge("서브주제 11", "하위주제 11.1")

# 노드들의 위치를 설정 (더 복잡한 레이아웃)
pos = nx.spring_layout(G, k=0.3, iterations=50)  # 더 세밀한 배치

# 노드, 엣지, 레이블 그리고 제목 그리기
plt.figure(figsize=(15, 12))
nx.draw(G, pos, with_labels=True, node_size=3500, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray")

# 그래프 타이틀
plt.title("마인드맵", fontsize=15)

# 그래프 출력
plt.show()
```
